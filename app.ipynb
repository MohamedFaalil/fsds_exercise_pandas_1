```python
from utilities.request_utility import RequestUtility
import pandas as pd
```

## set data URL

```python
DATA_URL = "https://api.github.com/repos/pandas-dev/pandas/issues"
```

##  getting json content

```python
try:
    json_contents = RequestUtility.getJsonContentByRequesting(DATA_URL)
except Exception as ex:
    print(ex)
```

## checking json content by printing

```python
print(json_contents)
```

## making json content to data frame

```python
main_df = pd.DataFrame(json_contents, columns=['id', 'url', 'repository_url', 'labels_url', 'user'])
```

## manipulating user data

```python
lst = []
for i, c in enumerate(main_df['id']):
    dict = {
        'id': main_df['id'][i],
        'url': main_df['url'][i],
    }
    user_dict = main_df['user'][i]
    user_attrs_list = user_dict.keys()
    for user_attr in user_attrs_list:
        dict['user_' + str(user_attr)] = user_dict[user_attr]
    lst.append(dict)   
```

## checking manipulated data by printing

```python
lst
```

## manupilated data frame making

```python
manupilated_lst_to_data_frame = pd.DataFrame(lst)
```

## csv file writing

```python
FILE_NAME = 'results.csv'
```

```python
manupilated_lst_to_data_frame.to_csv(FILE_NAME, index=None)
```
